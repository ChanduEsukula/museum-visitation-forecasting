Notebook to create the model dataset with lag-features by Andrew Miller

#1 Setup Environment, Load Libraries, and Load Merged Final Dataset

```python
# Load Libraries
import pandas as pd
from pathlib import Path

# Set data directory
DATA_DIR = Path("../data/processed")

# Load the final dataset to verify it was saved correctly
df = pd.read_csv(DATA_DIR / "final_dataset.csv")
df.head()
```

#2 Sort the Data by Month

```python
# Sort the DataFrame by month to ensure it's in the correct order
df = df.sort_values("month").reset_index(drop=True)
```

#3 Create the Lag Features

```python
# Create lag features and rolling mean
df["visitors_lag1"] = df["visitors"].shift(1)
df["visitors_lag12"] = df["visitors"].shift(12)
df["rolling_mean_3"] = df["visitors"].shift(1).rolling(3).mean()

#Inspect the lag features and rolling mean
df[["month", "visitors", "visitors_lag1", "visitors_lag12", "rolling_mean_3"]].head(15)
```

#4 Drop Rows with Missing Values Caused by the Lag Features

```python
# Drop rows with NaN values (due to lag features) and reset index
model_df = df.dropna().reset_index(drop=True)
```

#5 Check the Model Dataset

```python
# Inspect the final model dataset
print("Model dataset shape:", model_df.shape)
model_df.head()
```

#6 Save the Model Dataset CSV File

```python
# Save the model dataset
model_df.to_csv(DATA_DIR / "model_dataset.csv", index=False)
print("Saved model_dataset.csv")
```
