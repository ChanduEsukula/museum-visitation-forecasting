Museum attendance data preparation for Avila Adobe.  
Portfolio-cleaned notebook for loading, validating, and saving monthly visitor data.

#1 Setup Environment, Load Libraries and Load Raw CSV

```python
# importing necessary libraries
import pandas as pd
from pathlib import Path

# setting up directories for raw and processed data
RAW_DIR = Path("../data/raw/museum")
PROCESSED_DIR = Path("../data/processed")

# creating directories if they don't exist
RAW_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# defining file paths for raw and processed data
raw_file = RAW_DIR / "avila_adobe_monthly_raw.csv"
processed_file = PROCESSED_DIR / "museum_monthly.csv"

# reading the raw data file into a pandas DataFrame
df = pd.read_csv(raw_file)
df.head()
```

#2 Inspect the Columns

```python
# displaying the columns and shape of the DataFrame
print(df.columns.tolist())
print(df.shape)
df.head()
```

#3 Fix the Month Format and Columns Kept

```python
# converting the 'month' column to datetime format and then back to string format in "YYYY-MM" format
df["month"] = pd.to_datetime(df["month"]).dt.strftime("%Y-%m")
df.head()
```

```python
# creating a new DataFrame with only the 'month' and 'visitors' columns for further analysis
museum = df[["month", "visitors"]].copy()
museum.head()
```

#4 Check for Duplicate Months and Missing Values

```python
# checking for duplicate months, missing values, and the number of rows in the DataFrame
print("Duplicate months:", museum["month"].duplicated().sum())
print("\nMissing values:")
print(museum.isna().sum())
print("\nRows:", len(museum))
```

#5 Sort by Month

```python
# sorting the DataFrame by the 'month' column and resetting the index
museum = museum.sort_values("month").reset_index(drop=True)
museum.head()
```

#6 Save the Processed File

```python
# saving the processed DataFrame to a new CSV file in the processed data directory
museum.to_csv(processed_file, index=False)
print(f"Saved processed museum file to: {processed_file}")
```
