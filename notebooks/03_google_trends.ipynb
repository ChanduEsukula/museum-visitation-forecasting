## What the trend numbers mean

Google Trends values are **normalized 0-100 within the selected run window and geography**.

That means:
- `100` = that term's highest observed search interest in this run
- `50` = about half of that term's peak within this run
- `0` = very low or no measurable interest in that month

These are **not** raw search counts.

### Important interpretation detail

The `*_raw.png` charts in this notebook still show Google Trends values on the direct 0-100 scale, but those are still relative Google Trends values, not literal counts of searches. The `*_scaled.png` charts go one step further and min-max scale the series to 0-1 so they can be visually compared to the black dashed Avila visitors line.

```python
import re
import shutil
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")

RAW_CSV = ROOT / 'data' / 'raw' / 'trends' / 'google-trends-marketing-v2' / 'trends_weekly_long.csv'
PLOTS_V5_NOTEBOOK = PLOTS_V5
MIN_SCALED_NONZERO_MONTHS = 6
MIN_SCALED_UNIQUE_VALUES = 3

TERM_COLORS = {
    "Olvera Street": "#1f77b4",
    "Los Angeles history": "#ff7f0e",
    "things to do in Los Angeles": "#2ca02c",
    "Avila Adobe": "#4d4d4d",
    "Getty Museum": "#9467bd",
    "La Brea Tar Pits": "#8c564b",
    "Griffith Observatory": "#e377c2",
    "things to do in LA for a couple hours": "#17becf",
    "things to do in LA for 2 hours": "#bcbd22",
    "things to do near Union Station Los Angeles": "#aec7e8",
    "things to do in downtown LA": "#d62728",
    "things to do in Los Angeles in winter": "#98df8a",
    "things to do in LA in December": "#ff9896",
    "things to do in LA in January": "#c5b0d5",
    "things to do in LA in February": "#c49c94",
    "indoor things to do in Los Angeles": "#f7b6d2",
    "family activities Los Angeles": "#dbdb8d",
    "kid friendly museums Los Angeles": "#9edae5",
    "family things to do in downtown LA": "#ad494a",
    "cultural heritage Los Angeles": "#8ca252",
    "heritage tourism Los Angeles": "#b5cf6b",
    "historic district Los Angeles": "#637939",
    "Mexican heritage Los Angeles": "#e7969c",
    "El Pueblo Los Angeles": "#7b4173",
    "cultural experiences Los Angeles": "#ce6dbd",
    "interactive museums Los Angeles": "#3182bd",
    "historical sites Los Angeles": "#fd8d3c",
    "historic attractions Los Angeles": "#fdae6b",
}
AVILA_LINE_STYLE = dict(color="black", linestyle="--", linewidth=1.8, alpha=0.85, label="Avila visitors")


def slugify(term: str) -> str:
    s = term.lower().strip()
    s = re.sub(r"[^a-z0-9]+", "_", s)
    return f"trends_{re.sub(r'_+', '_', s).strip('_')}"


def min_max_scale(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce")
    mn, mx = s.min(), s.max()
    if pd.isna(mn) or mn == mx:
        return pd.Series(0.0, index=s.index)
    return (s - mn) / (mx - mn)


def load_raw() -> pd.DataFrame:
    df = pd.read_csv(RAW_CSV)
    df["date"] = pd.to_datetime(df["date"], format="mixed")
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    return df


def load_visitors_final() -> pd.DataFrame:
    df = pd.read_csv(MUSEUM_CSV)
    out = df[["Month", "Avila Adobe"]].copy()
    out.columns = ["month", "avila_visitors"]
    out["month"] = pd.to_datetime(out["month"], errors="coerce").dt.to_period("M").astype(str)
    out["avila_visitors"] = pd.to_numeric(out["avila_visitors"], errors="coerce")
    return out


def merged_monthly_final(monthly: pd.DataFrame, visitors: pd.DataFrame):
    m = monthly.merge(visitors, on="month", how="left")
    x = pd.to_datetime(m["month"] + "-01", errors="coerce")
    return m, x


def visitor_window(m: pd.DataFrame, x: pd.Series):
    mask = m["avila_visitors"].notna()
    return m.loc[mask].copy(), x.loc[mask].copy()


def is_meaningful_scaled_series(vals: pd.Series) -> bool:
    s = pd.to_numeric(vals, errors="coerce")
    pos = s[s > 0]
    return len(pos) >= MIN_SCALED_NONZERO_MONTHS and pos.nunique() >= MIN_SCALED_UNIQUE_VALUES


def add_avila_scaled(ax, m, x):
    if "avila_visitors" in m.columns and m["avila_visitors"].notna().any():
        ax.plot(x, min_max_scale(m["avila_visitors"]), **AVILA_LINE_STYLE)


def save_plot(fig, name: str):
    path = PLOTS_V5_NOTEBOOK / name
    fig.savefig(path, dpi=160, bbox_inches="tight")
    plt.close(fig)
    print("Saved:", path)


def line_pair(m, x, term_col_pairs, title_scaled, title_raw, fname_scaled, fname_raw):
    m_scaled, x_scaled = visitor_window(m, x)
    available_scaled = []
    for label, col in term_col_pairs:
        if col not in m_scaled.columns:
            continue
        vals = pd.to_numeric(m_scaled[col], errors="coerce")
        if is_meaningful_scaled_series(vals):
            available_scaled.append((label, col))

    fig, ax = plt.subplots(figsize=(13, 6))
    for label, col in available_scaled:
        vals = pd.to_numeric(m_scaled[col], errors="coerce")
        scaled = min_max_scale(vals.where(vals > 0))
        ax.plot(x_scaled, scaled, label=label, linewidth=2, color=TERM_COLORS.get(label))
    add_avila_scaled(ax, m_scaled, x_scaled)
    ax.set_title(title_scaled, fontsize=13, fontweight="bold")
    ax.set_ylabel("Min-max scaled value (0-1)", fontsize=10)
    ax.set_ylim(-0.03, 1.03)
    if not x_scaled.empty:
        ax.set_xlim(x_scaled.min(), x_scaled.max())
        ax.xaxis.set_major_locator(mdates.YearLocator(1))
        ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    if not available_scaled:
        ax.text(0.5, 0.55, "Not enough nonzero months in the visitor window\nfor a meaningful normalized comparison.", ha="center", va="center", transform=ax.transAxes, fontsize=11)
    ax.legend(fontsize=8)
    fig.tight_layout()
    save_plot(fig, fname_scaled)

    available_raw = [(label, col) for label, col in term_col_pairs if col in m.columns and m[col].notna().any() and m[col].gt(0).any()]
    if not available_raw:
        return
    fig, ax = plt.subplots(figsize=(13, 6))
    for label, col in available_raw:
        ax.plot(x, pd.to_numeric(m[col], errors="coerce"), label=label, linewidth=2, color=TERM_COLORS.get(label))
    ax.set_title(title_raw, fontsize=13, fontweight="bold")
    ax.set_ylabel("Google Trends relative search interest (0-100)", fontsize=10)
    ax.set_ylim(0, 100)
    ax.legend(fontsize=8)
    fig.tight_layout()
    save_plot(fig, fname_raw)


def line_scaled_only(m, x, term_col_pairs, title, fname):
    m_scaled, x_scaled = visitor_window(m, x)
    available = []
    for label, col in term_col_pairs:
        if col not in m_scaled.columns:
            continue
        vals = pd.to_numeric(m_scaled[col], errors="coerce")
        if is_meaningful_scaled_series(vals):
            available.append((label, col))
    fig, ax = plt.subplots(figsize=(13, 6))
    for label, col in available:
        vals = pd.to_numeric(m_scaled[col], errors="coerce")
        scaled = min_max_scale(vals.where(vals > 0))
        ax.plot(x_scaled, scaled, label=label, linewidth=2, color=TERM_COLORS.get(label))
    add_avila_scaled(ax, m_scaled, x_scaled)
    ax.set_title(title, fontsize=13, fontweight="bold")
    ax.set_ylabel("Min-max scaled value (0-1)", fontsize=10)
    ax.set_ylim(-0.03, 1.03)
    if not x_scaled.empty:
        ax.set_xlim(x_scaled.min(), x_scaled.max())
        ax.xaxis.set_major_locator(mdates.YearLocator(1))
        ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
    if not available:
        ax.text(0.5, 0.55, "Not enough nonzero months in the visitor window\nfor a meaningful normalized comparison.", ha="center", va="center", transform=ax.transAxes, fontsize=11)
    ax.legend(fontsize=8)
    fig.tight_layout()
    save_plot(fig, fname)


def build_v5_plots_from_notebook():
    PLOTS_V5_NOTEBOOK.mkdir(parents=True, exist_ok=True)

    raw = load_raw()
    monthly = pd.read_csv(TRENDS_V2 / 'trends_monthly.csv')
    visitors = load_visitors_final()
    m, x = merged_monthly_final(monthly, visitors)

    for fname in ['top_keyword_opportunities.png', 'theme_opportunity_scores.png']:
        src = ROOT / 'outputs' / 'plots' / 'google_trends_marketing' / 'google-trends-marketing-v2' / fname
        if src.exists():
            shutil.copy2(src, PLOTS_V5_NOTEBOOK / fname)
            print('Copied:', fname)

    DATE_MIN = pd.Timestamp('2014-01-01')
    DATE_MAX = pd.Timestamp('2021-12-31')
    CMAP, VMIN, VMAX = 'viridis', 0, 100

    df_s = raw[(raw['date'] >= DATE_MIN) & (raw['date'] <= DATE_MAX)].copy()
    zero_terms = df_s.groupby('term')['value'].max()
    zero_terms = zero_terms[zero_terms == 0].index.tolist()
    df_s = df_s[~df_s['term'].isin(zero_terms) & (df_s['value'].fillna(0) > 0)]
    order = df_s.groupby('term')['value'].mean().sort_values(ascending=True).index.tolist()
    df_s['y'] = df_s['term'].map({t: i for i, t in enumerate(order)})

    fig, ax = plt.subplots(figsize=(14, len(order) * 0.42))
    sc = ax.scatter(df_s['date'], df_s['y'], c=df_s['value'], cmap=CMAP, vmin=VMIN, vmax=VMAX, s=df_s['value'] * 1.2 + 8, alpha=0.75, linewidths=0, edgecolors='none')
    fig.colorbar(sc, ax=ax, pad=0.01, label='Trend value (0-100)')
    ax.set_yticks(range(len(order)))
    ax.set_yticklabels(order, fontsize=8.5)
    ax.set_xlim(DATE_MIN, DATE_MAX)
    ax.xaxis.set_major_locator(mdates.YearLocator(1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.set_title('All Google Trends Terms — 2014 to 2021 (Dataset Period)', fontsize=13, fontweight='bold', pad=10)
    ax.set_xlabel('Month', fontsize=10)
    ax.set_ylabel('Search term', fontsize=10)
    ax.grid(axis='x', linestyle='--', alpha=0.3)
    fig.tight_layout()
    save_plot(fig, 'all_terms_scatter_2014_2021_by_term.png')

    selected = pd.read_csv(TRENDS_V2 / 'trends_selected_2014_2026.csv').rename(columns={'month': 'month_str'})
    selected_long = selected.melt(id_vars=['month_str'], var_name='col', value_name='value')
    selected_long['value'] = pd.to_numeric(selected_long['value'], errors='coerce')
    selected_long = selected_long[selected_long['value'].fillna(0) > 0].copy()
    slug_to_term = {
        slugify('Avila Adobe'): 'Avila Adobe',
        slugify('Getty Museum'): 'Getty Museum',
        slugify('La Brea Tar Pits'): 'La Brea Tar Pits',
        slugify('Griffith Observatory'): 'Griffith Observatory',
        slugify('Olvera Street'): 'Olvera Street',
        slugify('Los Angeles museums'): 'Los Angeles museums',
        slugify('things to do in LA'): 'things to do in LA',
        slugify('things to do in Los Angeles'): 'things to do in Los Angeles',
        slugify('Los Angeles attractions'): 'Los Angeles attractions',
        slugify('free museums Los Angeles'): 'free museums Los Angeles',
        slugify('historic attractions Los Angeles'): 'historic attractions Los Angeles',
        slugify('historical sites Los Angeles'): 'historical sites Los Angeles',
        slugify('things to do in downtown LA'): 'things to do in downtown LA',
        slugify('Los Angeles history'): 'Los Angeles history',
        slugify('life in the 1840s'): 'life in the 1840s',
        slugify('living history museum'): 'living history museum',
        slugify('historic house museum'): 'historic house museum',
        slugify('family activities Los Angeles'): 'family activities Los Angeles',
    }
    selected_long['term'] = selected_long['col'].map(slug_to_term)
    selected_long = selected_long.dropna(subset=['term'])
    selected_long['month'] = pd.to_datetime(selected_long['month_str'] + '-01', errors='coerce')
    order_2026 = selected_long.groupby('term')['value'].mean().sort_values(ascending=True).index.tolist()
    selected_long['y'] = selected_long['term'].map({t: i for i, t in enumerate(order_2026)})

    fig, ax = plt.subplots(figsize=(14, max(7, len(order_2026) * 0.48)))
    sc2 = ax.scatter(selected_long['month'], selected_long['y'], c=selected_long['value'], cmap=CMAP, vmin=VMIN, vmax=VMAX, s=selected_long['value'] * 1.25 + 10, alpha=0.8, linewidths=0, edgecolors='none')
    fig.colorbar(sc2, ax=ax, pad=0.01, label='Trend value (0-100)')
    ax.set_yticks(range(len(order_2026)))
    ax.set_yticklabels(order_2026, fontsize=8.5)
    ax.set_xlim(pd.Timestamp('2014-01-01'), pd.Timestamp('2025-12-31'))
    ax.xaxis.set_major_locator(mdates.YearLocator(1))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.set_title('Top Google Trends Terms — 2014 to 2026', fontsize=13, fontweight='bold', pad=10)
    ax.set_xlabel('Month', fontsize=10)
    ax.set_ylabel('Search term', fontsize=10)
    ax.grid(axis='x', linestyle='--', alpha=0.3)
    fig.tight_layout()
    save_plot(fig, 'top_terms_scatter_2014_2026.png')

    line_pair(m, x, [("Olvera Street", slugify("Olvera Street")), ("Los Angeles history", slugify("Los Angeles history")), ("things to do in Los Angeles", slugify("things to do in Los Angeles"))], 'Normalized Visitor vs Search-Interest Trends', 'Top Search-Interest Trends (Google Trends Relative Interest)', 'avila_vs_top_trends_scaled.png', 'avila_vs_top_trends_raw.png')
    line_scaled_only(m, x, [("Getty Museum", slugify("Getty Museum")), ("La Brea Tar Pits", slugify("La Brea Tar Pits")), ("Griffith Observatory", slugify("Griffith Observatory"))], 'Normalized Benchmark Searches vs Avila Visitors', 'avila_vs_benchmarks.png')
    line_pair(m, x, [("things to do in LA for a couple hours", slugify("things to do in LA for a couple hours")), ("things to do in LA for 2 hours", slugify("things to do in LA for 2 hours")), ("things to do near Union Station Los Angeles", slugify("things to do near Union Station Los Angeles")), ("things to do in downtown LA", slugify("things to do in downtown LA"))], 'Short Visit Intent Trends (Normalized)', 'Short Visit Intent Trends (Google Trends Relative Interest)', 'short_visit_intent_trends_scaled.png', 'short_visit_intent_trends_raw.png')
    line_pair(m, x, [("historical sites Los Angeles", slugify("historical sites Los Angeles")), ("historic attractions Los Angeles", slugify("historic attractions Los Angeles")), ("things to do in downtown LA", slugify("things to do in downtown LA"))], 'Extended Tourism Discovery Trends (Normalized)', 'Extended Tourism Discovery Trends (Google Trends Relative Interest)', 'tourism_discovery_extended_trends_scaled.png', 'tourism_discovery_extended_trends_raw.png')
    line_pair(m, x, [("interactive museums Los Angeles", slugify("interactive museums Los Angeles")), ("cultural experiences Los Angeles", slugify("cultural experiences Los Angeles"))], 'Engagement Strategy Trends (Normalized)', 'Engagement Strategy Trends (Google Trends Relative Interest)', 'engagement_strategy_trends_scaled.png', 'engagement_strategy_trends_raw.png')
    line_scaled_only(m, x, [("things to do in Los Angeles in winter", slugify("things to do in Los Angeles in winter")), ("things to do in LA in December", slugify("things to do in LA in December")), ("things to do in LA in January", slugify("things to do in LA in January")), ("things to do in LA in February", slugify("things to do in LA in February")), ("indoor things to do in Los Angeles", slugify("indoor things to do in Los Angeles"))], 'Seasonal Visit Intent Trends (Normalized)', 'seasonal_visit_intent_trends.png')
    line_scaled_only(m, x, [("family activities Los Angeles", slugify("family activities Los Angeles")), ("kid friendly museums Los Angeles", slugify("kid friendly museums Los Angeles")), ("family things to do in downtown LA", slugify("family things to do in downtown LA"))], 'Family Activity Trends (Normalized)', 'family_activity_trends.png')
    line_scaled_only(m, x, [("cultural heritage Los Angeles", slugify("cultural heritage Los Angeles")), ("heritage tourism Los Angeles", slugify("heritage tourism Los Angeles")), ("historic district Los Angeles", slugify("historic district Los Angeles")), ("Mexican heritage Los Angeles", slugify("Mexican heritage Los Angeles")), ("El Pueblo Los Angeles", slugify("El Pueblo Los Angeles")), ("cultural experiences Los Angeles", slugify("cultural experiences Los Angeles"))], 'Cultural Heritage Trends (Normalized)', 'cultural_heritage_trends.png')

    print('Finished rebuilding final v5 plots from the notebook.')


build_v5_plots_from_notebook()
```
